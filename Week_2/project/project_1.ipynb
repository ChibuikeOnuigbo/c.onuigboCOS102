START
INPUT items_purchased
SET total_cost = 0
FOR EACH item IN items_purchased:
    total_cost = total_cost + item_price
OUTPUT "Total Amount Due: ", total_cost
INPUT payment_amount
IF payment_amount >= total_cost THEN
    SET change = payment_amount - total_cost
    OUTPUT "Payment successful. Change: ", change
    OUTPUT "Print Receipt"
ELSE
    OUTPUT "Insufficient funds. Please provide more money."
ENDIF
STOP
