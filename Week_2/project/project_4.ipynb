```python
def calculate_atr():
    print("--- Izifin Technology ATR Calculator ---")
    
    try:
        exp = float(input("Enter staff years of experience: "))
        age = float(input("Enter staff age: "))
    except ValueError:
        print("Please enter valid numbers.")
        return

    # Check conditions based on project instructions
    if exp > 25 and age >= 55:
        atr = 5600000
    elif exp > 20 and age >= 45:
        atr = 4480000
    elif exp > 10 and age >= 35:
        atr = 1500000
    elif exp < 10 and age < 35:
        atr = 550000
    else:
        # Handling edge cases not explicitly defined in the prompt image
        atr = "Condition not met by standard tiers. Requires manual review."

    if isinstance(atr, int):
        print(f"\nThe Annual Tax Revenue (ATR) for this staff is: N{atr:,.2f}")
    else:
        print(f"\n{atr}")

# Run the function
calculate_atr()
```
