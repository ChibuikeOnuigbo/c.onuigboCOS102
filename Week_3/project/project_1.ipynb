```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd

# Load the employee data from Excel file
df = pd.read_excel('GIG-logistics.xlsx')

def show_department_members(name, department):
    # Create a new window
    window = tk.Toplevel(root)
    window.title("Welcome")
    window.geometry("500x300")

    # Welcome message
    welcome_label = tk.Label(window, text=f"Welcome {name}!")
    welcome_label.pack()

    # Get all employees in the same department
    dept_employees = df[df['Department'] == department]['Name'].tolist()
    # Done
    # Display department members
    members_label = tk.Label(window, text=f"Other members in {department} department:")
    members_label.pack()

    for employee in dept_employees:
        if employee != name:
            member_label = tk.Label(window, text=employee)
            member_label.pack()

def check_employee():
    name = name_entry.get()
    department = dept_entry.get()

    # Check if employee exists in the dataframe
    employee = df[(df['Name'] == name) & (df['Department'] == department)]

    if len(employee) > 0:
        show_department_members(name, department)
    else:
        messagebox.showinfo("Employee Status", f"Employee {name} in {department} department does not exist.")

# Create the main window
root = tk.Tk()
root.title("Employee Verification System")
root.geometry("500x200")

# Create name label and entry
name_label = tk.Label(root, text="Employee Name:")
name_label.pack()
name_entry = tk.Entry(root)
name_entry.pack()

# Create department label and entry
dept_label = tk.Label(root, text="Department:")
dept_label.pack()
dept_entry = tk.Entry(root)
dept_entry.pack()

# Create submit button
submit_button = tk.Button(root, text="Submit", command=check_employee)
submit_button.pack()

# Run the main event loop
root.mainloop()
```
