```python
def calculate_finances():
    while True:
        print("\n" + "="*30)
        print("Financial Calculator Menu:")
        print("1. Simple Interest")
        print("2. Compound Interest")
        print("3. Annuity Plan")
        print("4. Exit Program")
        print("="*30)
        
        choice = input("Enter your choice (1, 2, 3, or 4 to exit): ")
        
        # Exit condition
        if choice == '4':
            print("Exiting calculator. Goodbye!")
            break
            
        # Invalid choice handler
        if choice not in ['1', '2', '3']:
            print("Error: Invalid Selection. Please type 1, 2, 3, or 4.")
            continue

        # Get common inputs and check for negative/zero errors
        print("\n--- Enter Values ---")
        P_or_PMT = float(input("Enter Principal or Payment Amount: "))
        R_input = float(input("Enter Rate percentage (e.g., 5 for 5%): "))
        
        if P_or_PMT <= 0 or R_input <= 0:
            print("Error: Amounts and Rates must be greater than zero. Try again.")
            continue

        # Calculation 1: Simple Interest
        if choice == '1':
            T = float(input("Enter Time in years (T): "))
            if T <= 0:
                print("Error: Time must be greater than zero.")
                continue
                
            # Formula: A = P(1 + (R/100)T)
            A = P_or_PMT * (1 + (R_input / 100) * T)
            print(f"\nResult: The Simple Interest Amount (A) is: {A:.2f}")

        # Calculation 2: Compound Interest
        elif choice == '2':
            n = int(input("Enter number of times compounded per year (n): "))
            t = float(input("Enter Time in years (t): "))
            if n <= 0 or t <= 0:
                print("Error: Time and compounding periods must be greater than zero.")
                continue
                
            # Formula: A = P(1 + R/n)^(nt)
            # We divide R_input by 100 here to convert the percentage to a decimal for the math
            R_decimal = R_input / 100
            A = P_or_PMT * ((1 + R_decimal / n) ** (n * t))
            print(f"\nResult: The Compound Interest Amount (A) is: {A:.2f}")

        # Calculation 3: Annuity Plan
        elif choice == '3':
            n = int(input("Enter number of times compounded per year (n): "))
            t = float(input("Enter Time in years (t): "))
            if n <= 0 or t <= 0:
                print("Error: Time and compounding periods must be greater than zero.")
                continue
                
            # Formula: A = PMT * [((1 + R/n)^(nt) - 1) / (R/n)]
            R_decimal = R_input / 100
            numerator = ((1 + R_decimal / n) ** (n * t)) - 1
            denominator = R_decimal / n
            A = P_or_PMT * (numerator / denominator)
            print(f"\nResult: The Annuity Plan Amount (A) is: {A:.2f}")

calculate_finances()
```
