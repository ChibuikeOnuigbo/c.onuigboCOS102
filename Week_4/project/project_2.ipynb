```python
# Cell 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cell 2: Load Data
# Ensure 'Titanic.csv' is in the same folder
df_titanic = pd.read_csv('Titanic.csv')

# Cell 3: Data Cleaning
# Fill missing Age values with the median (as suggested in the notes)
df_titanic['Age'] = df_titanic['Age'].fillna(df_titanic['Age'].median())

# Fill missing Embarked (port of embarkation) with the mode (most frequent)
df_titanic['Embarked'] = df_titanic['Embarked'].fillna(df_titanic['Embarked'].mode()[0])

# Drop 'Cabin' column because it has too many missing values to be useful
df_titanic.drop('Cabin', axis=1, inplace=True)

# Cell 4: Visualisation 1 - Boxplot (Checking for Outliers/Spread)
plt.figure(figsize=(10, 6))
# Boxplot comparing the age of those who survived vs those who didn't
sns.boxplot(data=df_titanic, x='Survived', y='Age', palette='Set2')
plt.title('Age Distribution by Survival Status')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()

# Cell 5: Visualisation 2 - Scatterplot (Relationships)
plt.figure(figsize=(10, 6))
# Scatterplot to look at the relationship between Age and Fare paid
sns.scatterplot(data=df_titanic, x='Age', y='Fare', hue='Survived', alpha=0.6)
plt.title('Relationship Between Age, Fare, and Survival')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()

# Cell 6: Visualisation 3 - Correlation Heatmap
plt.figure(figsize=(8, 6))
# Generating the correlation heatmap for the Titanic numerical data
titanic_corr = df_titanic.corr(numeric_only=True)
sns.heatmap(titanic_corr, annot=True, cmap='RdYlGn', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Titanic Features')
plt.show()
```
